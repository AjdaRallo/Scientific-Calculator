```python
import math
class Calculator():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
    def addition(num1, num2):
        print("Addition: ", num1+num2)
    def subtraction(num1, num2):
        print("Subtraction: ", num1-num2)
    def multiplication(num1, num2):
        print("Multiplication: ", num1*num2)
    def division(num1, num2):
        print("Division: ", num1/num2)
    def sqrt(num1):
        print("Sqrt: ", math.sqrt(num1))
    def cbrt(num1):
        print("Cbrt: ", math.cbrt(num1))
    def acos(num1):
        print("acos: ", math.acos(num1))
    def tanh(num1):
        print("tanh: ", math.tanh(num1))
    def sinh(num1):
        print("sinh: ", math.sinh(num1))
    def degrees(num1):
        print("Degrees: ", math.degrees(num1))
    def radians(num1):
        print("Radians: ", math.radians(num1))
    def remainder(num1):
        print("Remainder: ", math.remainder(num1))
            
while True:
    print("Select a number: ")
    print("1: Addition")
    print("2: Subtraction")
    print("3: Multiplication")
    print("4: Division")
    print("5: Sqrt")
    print("6: Cbrt")
    print("7: acos")
    print("8: tanh")
    print("9: sinh")
    print("10: degrees")
    print("11: radians")
    print("12: remainder")
    print("13: Exit")
    choice=int(input("Enter your choice (1-13):"))
        
    if choice==1:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Calculator.addition(num1, num2)
    elif choice==2:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Calculator.subtraction(num1, num2)
    elif choice==3:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Calculator.multiplication(num1, num2)
    elif choice==4:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))    
        if num2==0:
            print("Infinity")
        else:
            Calculator.division(num1, num2)
    elif choice==5:
        num1=int(input("Enter the number to sqrt it: "))
        Calculator.sqrt(num1)
    elif choice==6:
        num1=int(input("Enter the number to cbrt it: "))
        Calculator.cbrt(num1)
    elif choice==7:
        num1=int(input("Enter the number to acos it: "))
        Calculator.acos(num1)
    elif choice==8:
        num1=int(input("Enter the number to tanh it: "))
        Calculator.tanh(num1)
    elif choice==9:
        num1=int(input("Enter the number to sinh it: "))
        Calculator.sinh(num1)
    elif choice==10:
        num1=int(input("Enter the number to give the degrees: "))
        Calculator.degrees(num1)
    elif choice==11:
        num1=int(input("Enter the number to give the radians"))
        Calculator.radians(num1)
    elif choice==12:
        num1=int(input("Enter the number to give a remainder"))
        Calculator.remainder(num1)
    elif choice==13:
        break
    else:
        print("Wrong number, enter a number between 1-5")
```
